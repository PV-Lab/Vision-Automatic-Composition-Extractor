```python
# Import
import vision
import compextractor
import os
import pandas as pd
```

```python
# Variables
rasterSpeed = 38 # mm/s gocde raster speed
rasterScaleY = 0.9 # scale gcode raster pattern in Y-dim
rasterOffsetX = -45 # offset gcode raster pattern in X-dim
rotate_crop_params = {'theta': -0.5, 'x1': 45, 'x2': 830, 'y1': 120, 'y2': 550} # USER DEFINED ROTATE/CROP PARAMETERS FOR IMAGE
```

```python
# Read motor speed and gcode files
datapath = os.getcwd()+'\\data\\'
with open(datapath+'motor_speeds.txt', 'r') as file:
    motor_speeds = file.read()
gcode = pd.read_csv(datapath+'gcode_XY.csv')

# Run vision segmentation on RGB image
sgmnt, _ = vision.segmentation(img=datapath+'MAFA_RGB_Image.png', rotate_crop_params=rotate_crop_params)
```

```python
# Extract composition of segmented data using motor speeds and gcode
compextractor.get_compositions(segmentation=sgmnt, motor_speeds=motor_speeds, gcode=gcode, rasterSpeed=38, rasterScaleY = 0.9, rasterOffsetX = -45)
```
